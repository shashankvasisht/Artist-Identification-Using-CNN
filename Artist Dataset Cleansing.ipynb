```python
gp = df.groupby(df['artist']).size()
gp = gp.sort_values(axis=0,ascending=False)
gp 
```

```python
top_artist_keys = gp[:9].index.values
top_artist_keys = top_artist_keys.tolist()
pprint(top_artist_keys)
```

```python
dt = df[df['artist'].isin(top_artist_keys)]  ### dataframe of only those top 9 artists....now we can map these into classes
dt
```

```python
artist_names = ['Pierre-Auguste Renoir','Boris Kustodiev','Jean-Baptiste-Camille Corot'
,'Claude Monet','Pyotr Konchalovsky','Ilya Repin','Camille Pissarro','Martiros Saryan','Paul Gauguin']
print type(artist_names)
```

```python
artist_mapping = dict(zip(top_artist_keys,artist_names)) ## created a mapping dictionary for all artists...thus converting such a long string to a number..ie class index
print artist_mapping

```

```python
print artist_mapping['121fffad1eb6f7dff228b8a71b6aec72']
```

```python
#mapping it onto the dataframe
dt.artist = dt.artist.map(artist_mapping)
dt
```
